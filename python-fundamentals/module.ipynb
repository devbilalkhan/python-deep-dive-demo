`globals()` Python mein ek built-in function hai jo aapko current global symbol table ko dictionary ke form mein wapas karta hai. Global symbol table, wo jagah hai jahan Python apne variables aur objects ko store karta hai jo program ke global scope mein defined hote hain. Yani ke, agar aap kisi variable ya function ko program ke kisi bhi hisse se access kar sakte hain, to wo global namespace mein exist karta hai.

Jab aap `globals()` function ko call karte hain, to ye aapko ek dictionary return karta hai jismein key-value pairs hote hain. Har key ek global variable ya function ka name hota hai, aur corresponding value us variable ya function ki value hoti hai. Iska istemal aksar debugging ke doran ya runtime par program ke global state ko dekhne ke liye kiya jata hai.

```python
# Her python module ka aik global scope hota hai. 

globals()
```

```python
# Ye line 'math' module ko global namespace se access kar rahi hai

globals()['math']
```

```python
# Ye line 'math' module ki memory address ko hexadecimal form mein print kar rahi hai

hex(id(math))
```

```python
# math module ka naam return karta hai

math.__name__
```

```python
# math module ki dictionary, jisme iske attributes aur methods hote hain, return karta hai


math.__dict__ 
```

```python
# math module ki file location return karta hai

math.__file__
```
