```python
from pathlib import Path
from zipfile import ZipFile
```

```python
# yeh function zip file ko unzip ketha hao aur 
# output dir may safe kerta hai 

def extract_zip_data(path):
    """
    Extract zipfiles into output directory
    """
    output_path = Path('/kaggle/working/') / 'raw'
    output_path.mkdir(parents=True, exist_ok=True)
    with ZipFile(file=path) as f:
        f.extractall(path=output_path)
```

```python
# unzip and save in raw directory

extract_zip_data('/kaggle/input/nyc-taxi-trip-duration/train.zip')
```
